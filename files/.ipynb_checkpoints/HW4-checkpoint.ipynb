# 1. MIN TRANSFORMATION

```python
min_transform_df_temp = df.copy()
```

```python
min_transform_df = min_transform_df_temp.groupby('Bag Id').apply(np.min)
min_transform_df.head()
```

```python
X = min_transform_df.iloc[:,2:]
y = min_transform_df['Bag class']
```

### Random Forest

```python
scores = {}
n_trees = np.arange(1,100)
for i in n_trees:
    rf_base = RandomForestClassifier(random_state = 582, n_estimators=i)
    scores[i] = cross_val_score(rf_base, X, y, scoring='accuracy', cv=3).mean()
    
sns.lineplot(x= scores.keys(), y= scores.values());
```

```python
### After n=50, accuracy doesn't improve that much, therefore I will use values between 50 and 75, to be on the safe side.
```

```python
rf_base = RandomForestClassifier(random_state = 582)

param_grid = {'n_estimators': np.arange(50, 75),
             'max_features': np.arange(5, 20), # sqrt(166)~13, therefore try values around this.
             'max_depth': np.arange(5,100),
             'min_samples_leaf': np.arange(1,10)
             }

rf_cv = RandomizedSearchCV(rf_base, param_grid, n_iter=100, scoring='accuracy', cv=10, verbose=1, random_state=582)

rf_tuned = rf_cv.fit(X, y)
```

```python
print(f"Best Score: {rf_tuned.best_score_:.4f}")
print("Best Set of Params of RF for Min Transform: ",rf_tuned.best_params_)
```

### SVM Classifier

```python
svc_base = SVC(random_state = 582)

param_grid_svc = {'C': [0.0001, 0.001, 0.01, 0.1, 1, 5, 10, 100],
             'kernel': ['linear','poly','rbf'],
             'degree': [5,6,7]}

svc_cv = GridSearchCV(svc_base, param_grid_svc, scoring='accuracy', cv=10, verbose=1)
svc_tuned = svc_cv.fit(X, y)
```

```python
print(f"Best Score: {svc_tuned.best_score_:.4f}")
print("Best Set of Params of SVC for Min Transform: ",svc_tuned.best_params_)
```

```python
# SVM was really fast and has a superior output than of the RF.
```

# 2. HAUSDORFF DISTANCE

```python
# https://arxiv.org/pdf/1309.5643.pdf

# In this paper, bag dissimarity representation methods are proposed for the MIL problems, and seems to give good results.
# Therefore, I will be trying one of the proposed dissimilarity approaches, namely 'Hausdorff Distance'.It makes sense that
# bags that are closer in terms of this distance will be more likely to belong to same class. The distance is defined as
# such: "Hausdorff Distance between two point sets is the largest of the distances between each possible pair (a,b), such 
# that 'a' belongs to the first set and 'b' belongs to the second set.".
```

```python
new_df = df.copy()
new_df.head()
```

```python
# Pick the first bag from the dataset, that has class=1
a = new_df[new_df["Bag Id"]==1].iloc[:,2:].copy()

# Select the bags that belong to class1, and bags that belong to class 2.

class1_dict = {}
class2_dict = {}

for i in range(1,48):
    class1_dict[i] = new_df[new_df["Bag Id"]==i].iloc[:,2:].to_numpy()
    
for i in range(48,93):
    class2_dict[i] = new_df[new_df["Bag Id"]==i].iloc[:,2:].to_numpy()
```

```python
# Calculate the Hausdorff distances from the bag with id=1, to the other bags.

class1_len = []
for i in class1_dict.values():
    class1_len.append(directed_hausdorff(i, a)[0])
    
class2_len = []
for i in class2_dict.values():
    class2_len.append(directed_hausdorff(i, a)[0])
```

```python
np.mean(class1_len)
```

```python
np.mean(class2_len)
```

```python
# It seems like bags that have the same class, in average, has a lower hausdorff distance to the bag1 . Therefore, they are
# more similar, (less dissimilar) to the bag1.
```

### Let's convert all the points to the Hausdorff distance.

```python
# Merge class1 and class2 point sets
z = {**class1_dict, **class2_dict}
```

```python
# Create the new feature matrix, a distance-matrix, by creating distance vectors for each bag (using Hausdorff distance of
# that bag to the other bags).

new_distances = {}
for i in range(1,93):
    distances = []
    for j in range(1,93):
        distances.append(directed_hausdorff(new_df[new_df["Bag Id"]==i].iloc[:,2:],
                                            new_df[new_df["Bag Id"]==j].iloc[:,2:])[0])
    new_distances[f"bag{i}"] = distances
```

```python
# Specify the Input and Target features

new_features = pd.DataFrame(new_distances).T
classes = df.groupby("Bag Id")["Bag class"].mean()
```

### SVM Classifier

```python
svc_base_2 = SVC(random_state = 582)

param_grid_svc_2 = {'C': [8, 9, 10],
             'kernel': ['poly', 'rbf'],
             'degree': [5],
             'gamma': ['scale', 'auto']}

svc_cv_2 = GridSearchCV(svc_base_2, param_grid_svc_2, scoring='accuracy', cv=10, verbose=1)
svc_tuned_2 = svc_cv_2.fit(new_features, classes)
```

```python
print(f"Best Score: {svc_tuned_2.best_score_:.4f}")
print("Best Params: ",svc_tuned_2.best_params_)
```

```python
# As we can see, this bag representation is really good, compared to the previous one. I will use random forest, again, as
# my second model, just because I want to compare it with the previous RF model, that used MinTransform.
```

```python
rf_base_2 = RandomForestClassifier(random_state = 582)

param_grid_2 = {'n_estimators': np.arange(50, 75),
             'max_features': np.arange(5, 20), # sqrt(166)~13, therefore try values around this.
             'max_depth': np.arange(5,100),
             'min_samples_leaf': np.arange(1,10)
             }

rf_cv_2 = RandomizedSearchCV(rf_base_2, param_grid_2, n_iter=100, scoring='accuracy', cv=10, verbose=1, random_state=582)

rf_tuned_2 = rf_cv_2.fit(new_features, classes)
```

```python
print(f"Best Score: {rf_tuned_2.best_score_:.4f}")
print("Best Params: ",rf_tuned_2.best_params_)
```
